## Read data into np array

Express the cars.data.csv data as a series of histograms. The x-axis represents a feature and the y-axis is the frequency in the sample. Do this with the 'buying', 'maint', 'safety', and 'doors' fields with one plot for each for a total of four. Make each graph a subplot of a single output.

```python
import hw6

#Initialize Lists
car_data = hw6.CarEvaluationDataSet()
bad_data = []

#Get file name
# file_name = hw6.get_file_name()
file_name = "cars.data.csv"

#open and read file line by line. store variables. store any unreadable lines. report on num lines read and
#also show lines not read.
hw6.read_file(file_name, car_data, bad_data)
```

```python
from collections import defaultdict
# summarize data
buying = defaultdict(int)
maint = defaultdict(int)
doors = defaultdict(int)
safety = defaultdict(int)

for x in car_data.array[:,0]: buying[x] += 1
for x in car_data.array[:,1]: maint[x] += 1
for x in car_data.array[:,2]: doors[x] += 1
for x in car_data.array[:,5]: safety[x] += 1
```

```python
fig = pylab.figure()

# Buying
x = np.arange(len(buying))
ax = fig.add_subplot(221)
ax.set_title("Car Ratings: Price")
sorted_keys = sorted(buying.keys(), key=lambda i: hw6.CarEvaluationDataSet.buyingKeys[i])
bar_heights = [buying[i] for i in sorted_keys]
pylab.bar(x, bar_heights, align='center', width=0.5)
pylab.xticks(x, sorted_keys)
ymax = 1.05 * max(buying.values())
pylab.ylim(0, ymax)

# Maint
x = np.arange(len(maint))
ax = fig.add_subplot(222)
ax.set_title("Car Ratings: Maintainability")
sorted_keys = sorted(maint.keys(), key=lambda i: hw6.CarEvaluationDataSet.maintKeys[i])
bar_heights = [maint[i] for i in sorted_keys]
pylab.bar(x, bar_heights, align='center', width=0.5)
pylab.xticks(x, sorted_keys)
ymax = 1.05 * max(maint.values())
pylab.ylim(0, ymax)

# Doors
x = np.arange(len(doors))
ax = fig.add_subplot(223)
ax.set_title("Car Ratings: Number of Doors")
sorted_keys = sorted(doors.keys(), key=lambda i: hw6.CarEvaluationDataSet.doorsKeys[i])
bar_heights = [doors[i] for i in sorted_keys]
pylab.bar(x, bar_heights, align='center', width=0.5)
pylab.xticks(x, sorted_keys)
ymax = 1.05 * max(doors.values())
pylab.ylim(0, ymax)

# Safety
x = np.arange(len(safety))
ax = fig.add_subplot(224)
ax.set_title("Car Ratings: Safety Factor")
sorted_keys = sorted(safety.keys(), key=lambda i: hw6.CarEvaluationDataSet.safetyKeys[i])
bar_heights = [safety[i] for i in sorted(safety.keys())]
pylab.bar(x, bar_heights, align='center', width=0.5)
pylab.xticks(x, sorted_keys)
ymax = 1.05 * max(safety.values())
pylab.ylim(0, ymax)

# Spacing
fig.tight_layout()
```

# 2. Linear Regression

Plot your results from the linear regression in homework 5 and 7. The plot should include.  
 1) a scatter of the points in the brainandbody.csv.  
 2) a line showing the regression line (either from the calculation in HW5 or line-fitting from HW7).  
 3) something on the plot that specifies the equation for the regression line.  

```python
import hw7
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

def linear_func(x, m, b):
    return m * x + b
```

```python
brain_and_body = hw7.load_file_with_numpy("brainandbody.csv")
popt, pcov = curve_fit(linear_func, brain_and_body[:, 1], brain_and_body[:, 2])
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(brain_and_body[:, 1], brain_and_body[:, 2])

x = linspace(-500, 1.1 * max(brain_and_body[:,1]), 100)
y = linear_func(x, popt[0], popt[1])

ax.plot(x, y, 'r--', label='line eq')

# Axes
ax.set_xlim(0, 7000)
ax.set_ylim(0, 7000)

# Annotations
ax.annotate('Brain Weight = 6654\n'
            'Body Weight = 5712\n'
            'What is this big?\n'
            'What scale is this?\n'
            'How can the brain\n'
            'weigh more than the body?', 
            xy=(6654, 5712), 
            xytext=(4500, 4000),
            arrowprops=dict(arrowstyle='->'), 
            ha='left', va='top')
ax.annotate('This point is very large\n'
            'Compared to all the rest', 
            xy=(2547, 4603), 
            xytext=(1000, 6000),
            arrowprops=dict(arrowstyle='->'), 
            ha='left', va='top')
ax.annotate('Equation of Line: Y = {0:0.2f}*X + {1:0.2f}'.format(popt[0], popt[1]), 
            xy=(3000, 1000))

# Titles
ax.set_title('Body Weight vs Brain Weight')
ax.set_xlabel('Brain Weight')
ax.set_ylabel('Body Weight')

plt.show()
```
